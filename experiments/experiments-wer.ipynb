```python
###english
y_en = dev_wer_info.loc[(dev_wer_info['lang']=='eng') & (dev_wer_info['asr']=='en-US'),['wer','dialect']]
y_en_v = dev_wer_info.loc[(dev_wer_info['lang']=='eng') & (dev_wer_info['asr']!='en-US'),['wer','dialect']]
d_dict = dict(zip(y_en_v['dialect'],y_en_v['wer']))
y_en['us_wer']=y_en['wer']
y_en.drop(columns='wer', inplace=True)
y_en['v_wer']=y_en.apply(lambda x: d_dict[x['dialect']] if x['dialect'] in d_dict.keys() else 0,axis=1)

###ara
y_ar = dev_wer_info.loc[(dev_wer_info['lang']=='ara'),['wer','dialect']]

###ben
y_ben = dev_wer_info.loc[(dev_wer_info['lang']=='ben'),['wer','dialect','asr']]
y_ben['bn-BD']=y_ben.apply(lambda x: x['wer'] if x['asr']=='bn-BD' else None, axis=1)
y_ben['bn-IN']=y_ben.apply(lambda x: x['wer'] if x['asr']=='bn-IN' else None, axis=1)

###swe
y_swe = dev_wer_info.loc[(dev_wer_info['lang']=='swa'),['wer','dialect','asr']]
y_swe['sw-KE']=y_swe.apply(lambda x: x['wer'] if x['asr']=='sw-KE' else None, axis=1)
y_swe['sw-TZ']=y_swe.apply(lambda x: x['wer'] if x['asr']=='sw-TZ' else None, axis=1)

###kor
y_kor = dev_wer_info.loc[(dev_wer_info['lang']=='kor'),['wer','dialect']]
```

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('seaborn')

fig=plt.figure(figsize=(15,12))
fig.show()
ax=fig.add_subplot(321)
# ax.set_ylim([0, 5])

y_en.plot.bar(ax=ax, x='dialect',y=['us_wer','v_wer'], rot=90, width = 0.5)
plt.legend(["en-US", "en-VAR"])
plt.ylabel('word error rate (wer)')
plt.xlabel('English dialect')
# plt.title('English')

ax=fig.add_subplot(322)
y_ar.plot.bar(ax=ax, x='dialect',y=['wer'], rot=90, width = 0.2)
plt.legend(["ara-VAR"],loc=1)
plt.ylabel('word error rate (wer)')
plt.xlabel('Arabic dialect')
# plt.title('Arabic')

ax=fig.add_subplot(323)
y_ben.plot.bar(ax=ax, x='dialect',y=['bn-BD', 'bn-IN'], mark_right=True, rot=90, width = 0.2)
plt.legend(loc='best')
plt.ylabel('word error rate (wer)')
plt.xlabel('Bengali dialect')
# plt.title('Bengali')

ax=fig.add_subplot(324)
y_swe.plot.bar(ax=ax, x='dialect',y=['sw-KE', 'sw-TZ'], mark_right=True, rot=90, width = 0.2)
plt.legend(loc='best')
plt.ylabel('word error rate (wer)')
plt.xlabel('Kiswahili dialect')

ax=fig.add_subplot(325)
y_kor.plot.bar(ax=ax, x='dialect',y=['wer'], mark_right=True, rot=90, width = 0.2)
plt.legend(["kor-VAR"],loc='best')
plt.ylabel('word error rate (wer)')
plt.xlabel('Korean dialect')

plt.tight_layout()
plt.draw()
```

### Compute wer for each questions

```python
from jiwer import wer
f1='dev'
wer_df = {}
for f2 in os.listdir(os.path.join(root_path,f1)):
    if f2 not in discard:
        print(lang_st[f2])
        all_data  = pd.read_csv(os.path.join(root_path,f1,f2,'{}.{}.csv'.format(lang_st[f2],f1)), sep=",")
        del all_data['Unnamed: 0']
        all_data.fillna("",inplace=True)
        all_data[lang_st[f2]+'-gold'] = all_data.apply(lambda row: row[lang_st[f2]+'-gold'][:-1],axis=1)
        all_data[lang_st[f2]+'-gold'] = all_data.apply(lambda row: row[lang_st[f2]+'-gold'][:-1]
                                                       if row[lang_st[f2]+'-gold'][-1]==' '
                                                       else row[lang_st[f2]+'-gold'],axis=1)
        print(all_data.shape)
        wer_df[f2] =  pd.DataFrame()
        wer_df[f2]['ID'] = all_data['ID']
        gold = all_data.columns[2]
        for col in all_data.columns[3:]:
            col_name = '{}'.format(col)
            wer_df[f2][col_name] = all_data.apply(lambda row: round(wer(row[gold].lower(), row[col].lower()),3)
                                              if row[col]!="" else 1, axis=1)
#             wer_df[f2].to_csv(os.path.join(root_path,f1,f2,'wer.{}.{}.csv'.format(lang_st[f2], f1)))
```
